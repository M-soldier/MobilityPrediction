```python
import math
import json
import matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
matplotlib.rc("font",family='DengXian')
```

```python
def visualizeResults(res_path, model):
    with open(res_path, "r") as f:
        load_dict = json.load(f)

    train_loss = load_dict["metrics"]["train_loss"]
    valid_loss = load_dict["metrics"]["valid_loss"]
    accuracy = load_dict["metrics"]["accuracy"]

    x = range(1, len(train_loss)+1)
    
    fig = plt.figure(figsize=(20,6))
    gs = fig.add_gridspec(1, 2)
    ax1 = fig.add_subplot(gs[0,0])
    ax2 = fig.add_subplot(gs[0,1])
    # ax1, ax2 = fig.subplots(1, 2, sharey=True)

    ax1.plot(x, train_loss, label="训练集损失函数值", marker=".")
    ax1.plot(x, valid_loss, label="测试集损失函数值", marker="1")
    ax1.set_xticks(range(0, len(train_loss)+1, 5))
    ax1.set_xlabel("训练批次", size=15)
    ax1.set_ylabel("损失函数值", size=15)
    ax1.set_title("损失函数曲线", size=20)
    ax1.legend()

    ax2.plot(x, accuracy, label="测试集正确率", marker=".")
    ax2.set_xticks(range(0, len(train_loss)+1, 5))
    ax2.set_ylim(0, 0.2)
    ax2.set_xlabel("训练批次", size=15)
    ax2.set_ylabel("正确率", size=15)
    ax2.set_title("预测正确率曲线", size=20)
    ax2.legend()

    fig.suptitle(model + "模型结果", size=30, y=1.05)
    
    plt.show()


    # print(load_dict["metrics"])
```

```python
visualizeResults("../results/res/res_DeepMove_GRU_2023-03-25 12_08.rs", "DeepMove")
visualizeResults("../results/res/res_RNN_GRU_2023-03-25 09_15.rs", "GRU")
visualizeResults("../results/res/res_RNN_LSTM_2023-03-25 13_29.rs", "LSTM")
visualizeResults("../results/res/res_RNN_RNN_2023-03-25 12_51.rs", "RNN")
```
